# Giải thích các bước tính toán và Phân lớp bằng SVM

SVM (Support Vector Machine - Máy học véc-tơ hỗ trợ) là một trong những thuật toán phân loại kinh điển và mạnh mẽ nhất trong Machine Learning. Nếu GLCM đóng vai trò "cặp mắt" để trích xuất đặc trưng, thì SVM chính là "bộ não" tiếp nhận các con số đó để vạch ra ranh giới quyết định cuối cùng.

Dưới đây là quy trình 6 bước chuẩn mực để xây dựng và tính toán phân lớp bằng SVM:

## Bước 1: Biểu diễn Dữ liệu và Phân tách Không gian (Data Representation & Splitting)

Mục tiêu của bước này là tước bỏ mọi định dạng văn bản thừa thãi, đưa dữ liệu vào không gian toán học và thiết lập nền tảng để đánh giá mô hình.

1.  **Biểu diễn không gian:** Ma trận đặc trưng $X$ (các thông số GLCM) và Vector nhãn $y$ được trích xuất từ file lưu trữ lên RAM. Mỗi mẫu vật (ví dụ: một bức ảnh) chính thức trở thành **một điểm tọa độ** trong không gian đa chiều.
2.  **Phân chia dữ liệu (Data Splitting):** Nhằm đánh giá khách quan khả năng tổng quát hóa của mô hình và tránh hiện tượng quá khớp (Overfitting), tập dữ liệu gốc được phân chia ngẫu nhiên và phân tầng (**Stratified split**) thành hai tập độc lập:
    * **Tập huấn luyện (Training Set):** Chiếm tỷ lệ 70% - 80%, dùng để thiết lập ranh giới quyết định.
    * **Tập kiểm thử (Testing Set):** Chiếm 20% - 30% còn lại, đóng vai trò là tập dữ liệu mù (unseen data) để kiểm nghiệm hiệu năng thực tế.



## Bước 2: Tìm kiếm các Ranh giới Phân tách (Candidate Hyperplanes)

Thuật toán bắt đầu tìm các phương án vẽ vạch kẻ phân ranh giới để tách biệt các nhóm dữ liệu trong tập huấn luyện.
* **Siêu phẳng (Hyperplane):** Một phương trình toán học dạng $w \cdot x + b = 0$ được đặt vào không gian. SVM sẽ thử vô số cách xoay, dịch chuyển ranh giới này để phân tách các lớp dữ liệu.
* **Rủi ro:** Một siêu phẳng "tốt" không chỉ cần phân tách đúng các điểm hiện có mà phải có độ an toàn cao để không phân loại nhầm các mẫu vật mới trong tương lai.

## Bước 3: Xác định Lề (Margin) và Véc-tơ Hỗ trợ (Support Vectors)

Để chọn ra ranh giới tối ưu nhất, SVM áp dụng một bộ quy tắc về vùng đệm an toàn.
* **Lề (Margin):** Là khoảng cách vuông góc từ đường ranh giới đến các điểm dữ liệu gần nhất của mỗi lớp.
* **Véc-tơ Hỗ trợ (Support Vectors):** Là những điểm dữ liệu nằm ngay sát mép của vùng lề. Chúng là những điểm "khó phân loại nhất" và mang tính quyết định tuyệt đối; nếu thay đổi vị trí của các điểm này, toàn bộ đường ranh giới sẽ thay đổi.



## Bước 4: Tối ưu hóa Toán học - Siêu phẳng biên tối đa (Maximum Margin)

Giai đoạn này máy tính thực hiện giải bài toán quy hoạch bậc hai để chốt lại phương trình chuẩn xác nhất.
* **Mục tiêu:** Tìm thông số $w$ và $b$ sao cho giá trị của **Lề (Margin) là lớn nhất**.
* **Ý nghĩa:** Một đường ranh giới nằm chính giữa khoảng trống, cách xa các nhóm dữ liệu nhất sẽ giúp mô hình có khả năng chịu lỗi và bền vững hơn trước các biến động của dữ liệu thực tế.



## Bước 5: Xử lý Dữ liệu Phi tuyến tính với "Kernel Trick"

Trong thực tế, dữ liệu thường đan xen lộn xộn và không thể chia tách bằng một đường thẳng phẳng.
* **Ánh xạ chiều:** SVM sử dụng các hàm toán học để đẩy dữ liệu lên một không gian có số chiều cao hơn (ví dụ từ 2D lên 3D). Tại đây, các điểm dữ liệu vốn dĩ đan xen sẽ tách rời nhau ra, cho phép một mặt phẳng cắt ngang qua dễ dàng.
* **Hàm RBF (Radial Basis Function):** Đây là hàm Kernel phổ biến nhất, giúp SVM có thể tạo ra các đường ranh giới uốn lượn phức tạp xung quanh các cụm dữ liệu mà không tốn quá nhiều tài nguyên tính toán.



## Bước 6: Đánh giá và Xây dựng Đường ống Suy luận (Inference Pipeline)

Đây là bước cuối cùng để đưa mô hình từ lý thuyết vào ứng dụng thực tiễn.
1.  **Đánh giá hiệu năng:** Mô hình học từ tập huấn luyện sẽ được đem đi "thi" trên **Tập kiểm thử**. Dựa vào kết quả này, ta tính toán các chỉ số như Accuracy, Precision, Recall và đặc biệt là **F1-Score** để đảm bảo mô hình không bị thiên vị.
2.  **Suy luận (Inference):** Sau khi đạt yêu cầu, mô hình được đóng gói thành một "đường ống" tự động. Khi có một ảnh mới:
    * *Ảnh $\rightarrow$ Trích GLCM $\rightarrow$ Chuẩn hóa Min-Max $\rightarrow$ Áp dụng phương trình SVM $\rightarrow$ Kết quả phân loại.*
    * Quá trình này diễn ra cực nhanh, cho phép ứng dụng trong các hệ thống nhận diện thời gian thực.

## Bước 1: Biểu diễn Dữ liệu và Phân tách Không gian (Data Representation & Splitting)

Mục tiêu của bước này là tước bỏ mọi định dạng văn bản thừa thãi, đưa dữ liệu vào không gian toán học và thiết lập nền tảng để đánh giá mô hình.

1.  **Biểu diễn không gian:** Ma trận đặc trưng $X$ (các thông số GLCM) và Vector nhãn $y$ được trích xuất từ file lưu trữ lên RAM. Mỗi mẫu vật (ví dụ: một bức ảnh) chính thức trở thành **một điểm tọa độ** trong không gian đa chiều.
2.  **Phân chia dữ liệu (Data Splitting):** Nhằm đánh giá khách quan khả năng tổng quát hóa của mô hình và tránh hiện tượng quá khớp (Overfitting), tập dữ liệu gốc được phân chia ngẫu nhiên và phân tầng (**Stratified split**) thành hai tập độc lập:
    * **Tập huấn luyện (Training Set):** Chiếm tỷ lệ 70% - 80%, dùng để thiết lập ranh giới quyết định.
    * **Tập kiểm thử (Testing Set):** Chiếm 20% - 30% còn lại, đóng vai trò là tập dữ liệu mù (unseen data) để kiểm nghiệm hiệu năng thực tế.

```python
# =====================================================================
# BƯỚC 1: BIỂU DIỄN DỮ LIỆU VÀ PHÂN TÁCH KHÔNG GIAN
# =====================================================================

# --- NẠP CÁC THƯ VIỆN CẦN THIẾT ---
import pandas as pd # Thư viện thao tác với dữ liệu dạng bảng (như Excel)
from sklearn.model_selection import train_test_split # Thư viện hỗ trợ chia tập dữ liệu
import matplotlib.pyplot as plt # Thư viện vẽ biểu đồ
import seaborn as sns # Thư viện vẽ biểu đồ (làm đẹp cho matplotlib)

# Thư viện mới để tạo Widget tương tác (Nút bấm, Menu chọn)
import ipywidgets as widgets
from IPython.display import display

print("🚀 Bắt đầu BƯỚC 1: Chuẩn bị dữ liệu và phân tách không gian...")

# ---------------------------------------------------------------------
# 1. ĐỌC DỮ LIỆU TỪ Ổ CỨNG LÊN RAM
# ---------------------------------------------------------------------
print("\n[1/4] Đang nạp dữ liệu chuẩn hóa từ file CSV...")
# Chiếc "xe tải" pandas sẽ chở toàn bộ dữ liệu từ ổ cứng lên RAM và lưu vào bảng df
df = pd.read_csv('data/glcm_features_vector_normalized.csv')

# ---------------------------------------------------------------------
# 2. PHÂN CHIA DỮ LIỆU (DATA SPLITTING)
# ---------------------------------------------------------------------
print("\n[2/4] Đang phân chia ngẫu nhiên và phân tầng (Train 80% - Test 20%)...")
# Thực hiện chia trên toàn bộ bảng (df) để giữ nguyên cấu trúc (tên lá, metadata) khi xuất file
df_train, df_test = train_test_split(
    df,
    test_size=0.2, # Dành ra 20% cho tập Test, 80% cho tập Train
    random_state=42, # Đặt seed để mỗi lần chạy code đều chia giống hệt nhau (dễ debug)
    stratify=df['Label_ID'] # Kỹ thuật phân tầng (Stratified split): Đảm bảo tỷ lệ các loại lá được chia đều
)

# ---------------------------------------------------------------------
# 3. XUẤT FILE CSV (TẬP HUẤN LUYỆN VÀ TẬP KIỂM THỬ ĐỘC LẬP)
# ---------------------------------------------------------------------
print("\n[3/4] Đang xuất 2 tập dữ liệu độc lập ra file CSV...")
train_file = 'dataset_train.csv'
test_file = 'dataset_test.csv'

# Xuất ra ổ cứng. Dùng utf-8-sig để đảm bảo font tiếng Việt (nếu có) không bị lỗi khi mở bằng Excel
df_train.to_csv(train_file, index=False, encoding='utf-8-sig')
df_test.to_csv(test_file, index=False, encoding='utf-8-sig')

print(f"  -> 💾 Đã lưu Tập huấn luyện (Training Set): {train_file} ({df_train.shape[0]} mẫu)")
print(f"  -> 💾 Đã lưu Tập kiểm thử  (Testing Set) : {test_file} ({df_test.shape[0]} mẫu)")

# ---------------------------------------------------------------------
# 4. BIỂU DIỄN KHÔNG GIAN: TƯỚC BỎ VĂN BẢN, TẠO MA TRẬN X VÀ VECTOR Y
# ---------------------------------------------------------------------
print("\n[4/4] Tước bỏ định dạng văn bản, đưa mẫu vật vào không gian toán học đa chiều...")

# Cột chứa text cần vứt bỏ để máy tính không báo lỗi (vì SVM chỉ hiểu các con số)
columns_to_drop = ['Label_ID', 'Ten_Tieng_Anh', 'Ten_Tieng_Viet', 'File_Name']
# Chỉ lọc những cột thực sự tồn tại trong file để tránh báo lỗi Code
columns_to_drop = [col for col in columns_to_drop if col in df.columns]

# --- CHO TẬP TRAIN ---
# Trích xuất Ma trận X (chỉ chứa số liệu GLCM)
X_train = df_train.drop(columns=columns_to_drop)
# Trích xuất Vector y (chứa Đáp án/Nhãn)
y_train = df_train['Label_ID']

# --- CHO TẬP TEST ---
X_test = df_test.drop(columns=columns_to_drop)
y_test = df_test['Label_ID']

print(f"  -> 🧮 Kích thước ma trận không gian Train (X_train): {X_train.shape}")
print(f"  -> 🧮 Kích thước ma trận không gian Test  (X_test) : {X_test.shape}")
print("\n✅ Hoàn tất Bước 1! Dữ liệu (X, y) đã sẵn sàng 100% để bơm vào mô hình SVM.")

# =====================================================================
# PHẦN KIỂM CHỨNG & HIỂN THỊ TRỰC QUAN (DEMO)
# =====================================================================
print("\n" + "="*50)
print("📊 BẮT ĐẦU HIỂN THỊ DEMO TRỰC QUAN DỮ LIỆU")
print("="*50)

# --- DEMO 1: KIỂM TRA "CƠ NGƠI" KHÔNG GIAN BẰNG CONSOLE ---
print("\n[DEMO 1] Kiểm tra kích thước không gian và Tọa độ mẫu đầu tiên:")
print(f"Kích thước ma trận X_train: {X_train.shape} -> Đã tạo thành công không gian {X_train.shape[1]} chiều!")
print(f"Kích thước vector y_train: {y_train.shape}")

print("\nThông tin của mẫu vật đầu tiên (Index 0) trong không gian:")
print(f"  -> Đáp án Nhãn (y): {y_train.iloc[0]}")
print(f"  -> Tọa độ {X_train.shape[1]}D (X): {X_train.iloc[0].values}")

# --- DEMO 2: VẼ BẢN ĐỒ PHÂN TÁN (CÓ TÍNH NĂNG CHỌN LABEL TƯƠNG TÁC) ---
print("\n[DEMO 2] Biểu đồ phân tán tương tác (Interactive Scatter Plot)...")
print("💡 MẸO: Giữ phím 'Ctrl' (hoặc 'Cmd' trên Mac) và click chuột để chọn/bỏ chọn nhiều loại lá cùng lúc.")

# Lấy danh sách tất cả các loại lá có trong tập dữ liệu (đã sắp xếp A-Z)
all_labels = sorted(df_train['Label_ID'].unique())

# Hàm cập nhật biểu đồ dựa trên lựa chọn của người dùng
def update_scatter_plot(selected_labels):
    plt.figure(figsize=(10, 6)) # Thiết lập khung tranh

    # Nếu người dùng bỏ chọn hết, hiển thị thông báo
    if len(selected_labels) == 0:
        plt.title("⚠️ Vui lòng chọn ít nhất 1 loại lá để hiển thị!", color='red')
        plt.axis('off')
        plt.show()
        return

    # Lọc dữ liệu chỉ giữ lại các loại lá được chọn
    filtered_df = df_train[df_train['Label_ID'].isin(selected_labels)]

    sns.scatterplot(
        x=filtered_df['Contrast'],  # Trục ngang X là Contrast
        y=filtered_df['Entropy'],   # Trục dọc Y là Entropy
        hue=filtered_df['Label_ID'], # Tô màu các điểm dựa theo loại lá
        palette='tab20',             # Bảng màu tab20 hỗ trợ lên tới 20 loại màu khác nhau
        hue_order=selected_labels,   # Cố định màu sắc không bị nhảy khi chọn ít label
        s=100                        # Kích thước của các chấm tròn
    )

    # Cố định trục tọa độ theo dữ liệu tổng để khi bật/tắt label biểu đồ không bị giật/zoom
    plt.xlim(df_train['Contrast'].min() - 0.05, df_train['Contrast'].max() + 0.05)
    plt.ylim(df_train['Entropy'].min() - 0.05, df_train['Entropy'].max() + 0.05)

    plt.title(f"Demo Không gian Đặc trưng (Đang hiển thị {len(selected_labels)} loại lá)")
    plt.xlabel("Trục Contrast")
    plt.ylabel("Trục Entropy")
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left') # Đẩy chú thích ra ngoài cho đỡ vướng đồ thị
    plt.grid(True)
    plt.tight_layout() # Căn chỉnh tự động để đồ thị không bị cắt xén
    plt.show() # Hiển thị hình ảnh

# Tạo Menu hộp chọn nhiều thành phần (SelectMultiple Widget)
label_selector = widgets.SelectMultiple(
    options=all_labels,
    value=all_labels, # Mặc định ban đầu là chọn tất cả
    description='🔍 Chọn Lá:',
    disabled=False,
    rows=min(8, len(all_labels)) # Hiển thị tối đa 8 dòng cuộn
)

# Kết nối Menu chọn với Hàm vẽ biểu đồ
widgets.interact(update_scatter_plot, selected_labels=label_selector)

# --- DEMO 3: KIỂM CHỨNG TÍNH CÔNG BẰNG (STRATIFIED SPLIT) ---
print("\n[DEMO 3] Đang vẽ biểu đồ cột kiểm tra tính cân bằng dữ liệu...")
# Đếm số lượng lá mỗi loại trong tập Train và tập Test
train_counts = y_train.value_counts().sort_index()
test_counts = y_test.value_counts().sort_index()

# Gom số liệu vào một DataFrame mới để dễ vẽ biểu đồ
demo_split = pd.DataFrame({
    'Tập huấn luyện (80%)': train_counts,
    'Tập kiểm thử (20%)': test_counts
})

# Lệnh vẽ biểu đồ cột (Bar chart)
demo_split.plot(kind='bar', figsize=(10, 5), color=['#1f77b4', '#ff7f0e'])
plt.title("Demo Phân bổ Dữ liệu sau khi chia Stratified Split")
plt.xlabel("Nhãn (Các loại lá)")
plt.ylabel("Số lượng mẫu")
plt.xticks(rotation=0) # Cho chữ ở trục X nằm ngang dễ đọc
plt.grid(axis='y') # Kẻ đường lưới nằm ngang
plt.show() # Hiển thị hình ảnh lên màn hình
```

## Bước 2: Tìm kiếm các Ranh giới Phân tách (Candidate Hyperplanes)

Thuật toán bắt đầu tìm các phương án vẽ vạch kẻ phân ranh giới để tách biệt các nhóm dữ liệu trong tập huấn luyện.

* **Siêu phẳng (Hyperplane):** Một phương trình toán học dạng $w \cdot x + b = 0$ được đặt vào không gian. Để hiểu cách SVM "dò đường", chúng ta có thể bóc tách phương trình này như sau:
    * **$x$ (Tọa độ dữ liệu):** Vị trí của các mẫu vật trong không gian (ví dụ: tọa độ trên trục Contrast và Entropy).
    * **$w$ (Weight - Trọng số):** Quyết định **góc xoay (độ nghiêng)** của ranh giới. Việc thuật toán thay đổi $w$ tương đương với hành động "xoay thước kẻ".
    * **$b$ (Bias - Độ chệch):** Quyết định **vị trí (sự tịnh tiến)** của ranh giới. Thay đổi $b$ giúp trượt thước kẻ lên/xuống/trái/phải trong không gian mà không làm thay đổi độ nghiêng.
    * **$= 0$:** Đại diện cho chính vạch kẻ ranh giới. SVM dựa vào đây để chấm điểm: Nếu thay tọa độ một chiếc lá vào phương trình mà ra kết quả $> 0$ thì nó thuộc lớp bên này, nếu $< 0$ thì nó thuộc lớp bên kia.

💡 *Như vậy, SVM thực chất đang chạy một vòng lặp khổng lồ, liên tục thay đổi các giá trị $w$ và $b$ (vừa xoay, vừa tịnh tiến) để gom lại tất cả các đường ranh giới lách qua được khe hở của 2 nhóm dữ liệu.*

* **Rủi ro:** Một siêu phẳng "tốt" không chỉ cần phân tách đúng các điểm hiện có mà phải có độ an toàn cao để không phân loại nhầm các mẫu vật mới trong tương lai. Nếu một đường ứng viên đi sát sạt vào mép của cụm dữ liệu, chỉ cần một chiếc lá thu thập ngày mai có hình dáng hơi dị biệt một chút là sẽ vô tình vượt biên và bị phân loại nhầm ngay.

```python
# =====================================================================
# BƯỚC 2: TÌM KIẾM CÁC RANH GIỚI PHÂN TÁCH (CANDIDATE HYPERPLANES)
# Đã fix lỗi Font chữ (Xóa Emoji trên biểu đồ Matplotlib)
# =====================================================================

# --- NẠP CÁC THƯ VIỆN CẦN THIẾT ---
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
from matplotlib.animation import FuncAnimation
from IPython.display import display, HTML, Markdown

# TĂNG GIỚI HẠN BỘ NHỚ CHO VIDEO LÊN 100MB
mpl.rcParams['animation.embed_limit'] = 100.0

# ---------------------------------------------------------------------
# 1. KHAI BÁO CẶP LÁ CẦN PHÂN TÁCH
# ---------------------------------------------------------------------
all_labels = sorted(df_train['Label_ID'].unique())
label_A = all_labels[0]
label_B = 'Leaf 9'

# ---------------------------------------------------------------------
# 2. THIẾT LẬP CỖ MÁY QUÉT & TÍNH TOÁN TOÁN HỌC
# ---------------------------------------------------------------------
if label_A == label_B:
    print("⚠️ LỖI: Vui lòng khai báo 2 loại lá KHÁC NHAU ở mục 1!")
else:
    df_demo = df_train[df_train['Label_ID'].isin([label_A, label_B])]
    X_demo = df_demo[['Contrast', 'Entropy']].values
    y_demo = df_demo['Label_ID'].values

    # BƯỚC TOÁN HỌC CỐT LÕI: TÌM TRỌNG TÂM VÀ ĐIỂM VÀNG
    mean_A = X_demo[y_demo == label_A].mean(axis=0)
    mean_B = X_demo[y_demo == label_B].mean(axis=0)
    cx, cy = (mean_A + mean_B) / 2

    global_x_min, global_x_max = df_train['Contrast'].min(), df_train['Contrast'].max()
    global_y_min, global_y_max = df_train['Entropy'].min(), df_train['Entropy'].max()
    y_range = global_y_max - global_y_min

    # ---------------------------------------------------------------------
    # 3. HIỂN THỊ TỪNG BƯỚC XÁC ĐỊNH ĐIỂM VÀNG BẰNG HÌNH ẢNH
    # ---------------------------------------------------------------------
    print("\n" + "="*70)
    print("🎬 GIAI ĐOẠN 1: MINH HỌA TỪNG BƯỚC XÁC ĐỊNH 'ĐIỂM VÀNG' ĐỂ LÀM TÂM XOAY")
    print("="*70)

    fig_steps, axes = plt.subplots(1, 3, figsize=(16, 5))

    # --- ĐỒ THỊ 1: DỮ LIỆU THÔ ---
    sns.scatterplot(x=X_demo[:, 0], y=X_demo[:, 1], hue=y_demo, palette=['#e74c3c', '#3498db'], s=80, alpha=0.7, ax=axes[0], legend=False)
    axes[0].set_title("1. Nhìn vào cụm dữ liệu thô", fontweight='bold')
    axes[0].set_xlim(global_x_min - 0.05, global_x_max + 0.05)
    axes[0].set_ylim(global_y_min - 0.05, global_y_max + 0.05)
    axes[0].grid(True)

    # --- ĐỒ THỊ 2: TÌM TRỌNG TÂM ---
    sns.scatterplot(x=X_demo[:, 0], y=X_demo[:, 1], hue=y_demo, palette=['#e74c3c', '#3498db'], s=80, alpha=0.3, ax=axes[1], legend=False)
    axes[1].scatter(mean_A[0], mean_A[1], color='darkred', marker='X', s=200, zorder=5)
    axes[1].scatter(mean_B[0], mean_B[1], color='darkblue', marker='X', s=200, zorder=5)
    axes[1].set_title("2. Tìm Trọng Tâm (Dấu X) của từng cụm", fontweight='bold')
    axes[1].set_xlim(global_x_min - 0.05, global_x_max + 0.05)
    axes[1].set_ylim(global_y_min - 0.05, global_y_max + 0.05)
    axes[1].grid(True)

    # --- ĐỒ THỊ 3: CHỐT ĐIỂM VÀNG ---
    sns.scatterplot(x=X_demo[:, 0], y=X_demo[:, 1], hue=y_demo, palette=['#e74c3c', '#3498db'], s=80, alpha=0.3, ax=axes[2])
    axes[2].scatter(mean_A[0], mean_A[1], color='darkred', marker='X', s=200, zorder=5)
    axes[2].scatter(mean_B[0], mean_B[1], color='darkblue', marker='X', s=200, zorder=5)
    axes[2].plot([mean_A[0], mean_B[0]], [mean_A[1], mean_B[1]], color='gray', linestyle='--', lw=2, zorder=4)
    axes[2].scatter(cx, cy, color='gold', marker='*', s=500, edgecolors='black', linewidths=1.5, zorder=6, label='Diem Vang (Tam xoay)')
    axes[2].set_title("3. Chốt Điểm Vàng nằm chính giữa", fontweight='bold')
    axes[2].set_xlim(global_x_min - 0.05, global_x_max + 0.05)
    axes[2].set_ylim(global_y_min - 0.05, global_y_max + 0.05)
    axes[2].legend(loc='lower right')
    axes[2].grid(True)

    plt.tight_layout()
    plt.show()

    # ---------------------------------------------------------------------
    # 4. CHUẨN BỊ QUÉT KHÔNG GIAN
    # ---------------------------------------------------------------------
    num_angles = 60
    num_offsets = 12
    angles = np.linspace(-np.radians(85), np.radians(85), num_angles)
    offsets = np.linspace(-y_range*0.4, y_range*0.4, num_offsets)

    total_lines = num_angles * num_offsets

    # Phần Markdown này trình duyệt hiển thị nên cứ giữ Emoji cho đẹp
    scan_info = fr"""
### 🎯 GIAI ĐOẠN 2: Bắt đầu dò đường quanh Điểm Vàng
* **Xoay quanh Điểm Vàng ($w$):** Quét qua **{num_angles}** góc khác nhau (từ $-85^\circ$ đến $+85^\circ$) với tâm quay là Ngôi sao 🌟.
* **Tịnh Tiến ($b$):** Ở mỗi góc, đường thẳng sẽ nhấc lên/hạ xuống qua **{num_offsets}** vị trí để vét cạn không gian.
* **Tổng cộng:** **{total_lines}** vạch kẻ phân cách sẽ được thử nghiệm!
    """
    display(Markdown(scan_info))
    print(f"⏳ Đang dựng Video phân tích {total_lines} đường ranh giới. Xin vui lòng chờ trong giây lát...")

    frames_data = []
    valid_candidates = []

    for angle in angles:
        for offset in offsets:
            m = np.tan(angle)
            intercept = cy + offset - m * cx

            side_A = (X_demo[y_demo == label_A][:, 1] - (m * X_demo[y_demo == label_A][:, 0] + intercept)) > 0
            side_B = (X_demo[y_demo == label_B][:, 1] - (m * X_demo[y_demo == label_B][:, 0] + intercept)) > 0

            is_valid = (np.all(side_A) and np.all(~side_B)) or (np.all(~side_A) and np.all(side_B))

            if is_valid:
                valid_candidates.append((m, intercept))

            frames_data.append((m, intercept, is_valid, angle, offset))

    # ---------------------------------------------------------------------
    # 5. DỰNG VIDEO BẢNG ĐIỆN TỬ CHẠY ANIMATION
    # ---------------------------------------------------------------------
    fig = plt.figure(figsize=(10, 7))
    gs = fig.add_gridspec(2, 1, height_ratios=[0.1, 0.9])

    # TẦNG TRÊN: BẢNG ĐIỆN TỬ
    ax_text = fig.add_subplot(gs[0])
    ax_text.axis('off')
    info_text = ax_text.text(0.5, 0.5, "", ha='center', va='center',
                             fontsize=13, fontweight='bold', color='#2c3e50',
                             bbox=dict(boxstyle="round,pad=0.4", facecolor="#f1f2f6", edgecolor="none"))

    # TẦNG DƯỚI: ĐỒ THỊ ANIMATION
    ax = fig.add_subplot(gs[1])
    sns.scatterplot(x=X_demo[:, 0], y=X_demo[:, 1], hue=y_demo, palette=['#e74c3c', '#3498db'], s=80, alpha=0.7, ax=ax)

    # Vẽ sẵn Điểm vàng và Trọng tâm
    ax.scatter(mean_A[0], mean_A[1], color='darkred', marker='X', s=200, zorder=5)
    ax.scatter(mean_B[0], mean_B[1], color='darkblue', marker='X', s=200, zorder=5)
    ax.plot([mean_A[0], mean_B[0]], [mean_A[1], mean_B[1]], color='gray', linestyle='--', lw=2, zorder=4)
    # Đã thay label bằng chữ thường
    ax.scatter(cx, cy, color='gold', marker='*', s=600, edgecolors='black', linewidths=1.5, zorder=6, label='* DIEM VANG')
    ax.legend(loc='upper right')

    line, = ax.plot([], [], lw=2, zorder=3)

    ax.set_xlim(global_x_min - 0.05, global_x_max + 0.05)
    ax.set_ylim(global_y_min - 0.05, global_y_max + 0.05)
    ax.set_xlabel("Trục Contrast (Tọa độ $x_1$)", fontweight='bold')
    ax.set_ylabel("Trục Entropy (Tọa độ $x_2$)", fontweight='bold')
    ax.grid(True)

    def init():
        line.set_data([], [])
        info_text.set_text("")
        return line, info_text

    def animate(i):
        m, intercept, is_valid, angle, offset = frames_data[i]
        x_vals = np.array([global_x_min - 1, global_x_max + 1])
        y_vals = m * x_vals + intercept

        if is_valid:
            line.set_color('#2ecc71')
            line.set_linewidth(5)
            status = "[OK] LUU LAI"  # Thay emoji bằng text
        else:
            line.set_color('#e74c3c')
            line.set_linewidth(2)
            status = "[X] BI LOAI"   # Thay emoji bằng text

        line.set_data(x_vals, y_vals)
        info_str = f"Góc xoay (w): {np.degrees(angle):>4.0f}°   |   Tịnh tiến (b): {offset:>5.2f}   |   Trạng thái: {status}"
        info_text.set_text(info_str)
        return line, info_text

    anim = FuncAnimation(fig, animate, init_func=init, frames=len(frames_data), interval=30, blit=True)
    plt.close(fig)

    display(HTML(anim.to_jshtml()))

    # ---------------------------------------------------------------------
    # 6. BỨC ẢNH TĨNH TỔNG HỢP SAU KHI QUÉT XONG
    # ---------------------------------------------------------------------
    print("\n" + "="*70)
    print(f"📸 GIAI ĐOẠN 3: TÌM THẤY {len(valid_candidates)}/{total_lines} ĐƯỜNG ỨNG VIÊN HỢP LỆ")
    print("="*70)

    if len(valid_candidates) > 0:
        fig2, ax2 = plt.subplots(figsize=(10, 6))
        sns.scatterplot(x=X_demo[:, 0], y=X_demo[:, 1], hue=y_demo, palette=['#e74c3c', '#3498db'], s=80, alpha=0.7, ax=ax2)

        ax2.scatter(mean_A[0], mean_A[1], color='darkred', marker='X', s=200, zorder=5)
        ax2.scatter(mean_B[0], mean_B[1], color='darkblue', marker='X', s=200, zorder=5)
        ax2.plot([mean_A[0], mean_B[0]], [mean_A[1], mean_B[1]], color='gray', linestyle='--', lw=2, zorder=4)
        ax2.scatter(cx, cy, color='gold', marker='*', s=600, edgecolors='black', linewidths=1.5, zorder=6)

        x_vals_static = np.array([global_x_min - 0.5, global_x_max + 0.5])
        for m, intercept in valid_candidates:
            y_vals_static = m * x_vals_static + intercept
            ax2.plot(x_vals_static, y_vals_static, color='#2ecc71', alpha=0.15, lw=2, zorder=3)

        ax2.set_xlim(global_x_min - 0.05, global_x_max + 0.05)
        ax2.set_ylim(global_y_min - 0.05, global_y_max + 0.05)
        ax2.set_title("Chùm Ranh giới Ứng viên (Cắt qua khe hở giữa 2 phe)", fontsize=14)
        ax2.set_xlabel("Trục Contrast (Tọa độ $x_1$)", fontweight='bold')
        ax2.set_ylabel("Trục Entropy (Tọa độ $x_2$)", fontweight='bold')
        ax2.grid(True)
        plt.show()

        print("💡 Chuẩn bị cho BƯỚC 3: Dựa vào chùm đường xanh lá phía trên, SVM sẽ tính toán Margin tới các Support Vectors để chọn ra 1 đường duy nhất an toàn nhất!")
    else:
        print("⚠️ Không tìm thấy đường ranh giới thẳng nào phân tách 100% hai loại lá này.")
```

## Bước 3: Xác định Lề (Margin) và Véc-tơ Hỗ trợ (Support Vectors)

Trong số hàng chục đường ranh giới (Candidate Hyperplanes) đã tìm thấy ở Bước 2, làm sao SVM chọn được một đường duy nhất "chuẩn" nhất? Thuật toán không chọn ngẫu nhiên mà sẽ thực hiện bài test đo đạc không gian thông qua 2 khái niệm sau:



* **1. Quá trình xác định Véc-tơ Hỗ trợ (Support Vectors):**
    * Từ một đường ranh giới, SVM sẽ bắt đầu "thổi phồng" một vùng đệm sang hai bên (giống như việc mở rộng một con đường). Nó cứ mở rộng dần đều cho đến khi vỉa hè "va chạm" vào những điểm dữ liệu đầu tiên của cả phe Xanh lẫn phe Đỏ.
    * Những điểm dữ liệu đầu tiên bị va chạm này chính là các **Véc-tơ Hỗ trợ (Support Vectors)**.
    * **Tầm quan trọng:** Chúng đóng vai trò như những "cột mốc" chốt chặn biên giới. Thuật toán SVM cực kỳ tối ưu vì nó **chỉ quan tâm đến những điểm này**. Dù phía sau chúng có hàng ngàn chiếc lá khác, thuật toán cũng sẽ phớt lờ đi, vì ranh giới đã được định hình hoàn toàn bởi các véc-tơ hỗ trợ nằm sát mép nhất. Nếu bạn dịch chuyển các điểm này, toàn bộ ranh giới sẽ bị lệch theo.

* **2. Quá trình xác định Lề (Margin) và Chốt hạ ranh giới:**
    * Sau khi vạch ra hai đường biên đi qua các Support Vectors, khoảng cách vuông góc đo được giữa hai đường biên này (bề rộng của toàn bộ con đường) được gọi là **Lề (Margin)**.
    * **Quy tắc tối hậu của SVM:** Thuật toán sẽ tính toán Lề cho *tất cả* các đường ranh giới ứng viên tìm được ở Bước 2. Đường nào tạo ra được **Lề rộng nhất (Maximum Margin)** sẽ được tuyên bố là người chiến thắng - Siêu phẳng Tối ưu (Optimal Hyperplane). Ranh giới càng rộng, mô hình càng phân loại an toàn và càng khó bị nhầm lẫn khi gặp dữ liệu mới trong thực tế!

```python
# =====================================================================
# BƯỚC 3: TÌM KIẾM ĐƯỜNG TỐI ƯU (MAX MARGIN) & SUPPORT VECTORS
# Cập nhật: Thêm chú thích (Legend) đầy đủ cho Video Giai đoạn 4
# =====================================================================

# ---------------------------------------------------------------------
# 1. BÀI TOÁN TOÁN HỌC: ĐO KHOẢNG CÁCH TỪ ĐIỂM ĐẾN ĐƯỜNG THẲNG
# ---------------------------------------------------------------------
if 'valid_candidates' not in locals() or len(valid_candidates) == 0:
    print("⚠️ LỖI: Không tìm thấy chùm đường ranh giới ứng viên. Vui lòng chạy lại Bước 2!")
else:
    intro_step3 = r"""
#### GIAI ĐOẠN 4: Quá trình "Thổi phồng" Lề (Margin) và tìm Support Vectors
Thuật toán quét lại từng đường ranh giới xanh lá để:
1. **Đo khoảng cách:** Tìm khoảng cách từ đường thẳng đến điểm gần nhất của mỗi phe.
2. **Xác định Support Vectors:** Khoanh tròn vàng các điểm "chạm" vào lề đầu tiên.
3. **Chọn Max Margin:** Đường nào có khoảng cách tới các Support Vectors lớn nhất sẽ thắng.
    """
    display(Markdown(intro_step3))
    print(f"⏳ Đang dựng Video phân tích chi tiết {len(valid_candidates)} đường ranh giới...")

    frames_step3 = []
    best_margin = -1
    best_candidate = None
    best_svs_A = None
    best_svs_B = None

    # XỬ LÝ TOÁN HỌC
    for m, intercept in valid_candidates:
        dists_A = np.abs(-m * X_demo[y_demo == label_A][:, 0] + X_demo[y_demo == label_A][:, 1] - intercept) / np.sqrt(m**2 + 1)
        dists_B = np.abs(-m * X_demo[y_demo == label_B][:, 0] + X_demo[y_demo == label_B][:, 1] - intercept) / np.sqrt(m**2 + 1)

        margin_A = np.min(dists_A)
        margin_B = np.min(dists_B)
        current_margin = min(margin_A, margin_B)

        sv_A = X_demo[y_demo == label_A][dists_A <= margin_A + 1e-5]
        sv_B = X_demo[y_demo == label_B][dists_B <= margin_B + 1e-5]

        frames_step3.append((m, intercept, current_margin, sv_A, sv_B))

        if current_margin > best_margin:
            best_margin = current_margin
            best_candidate = (m, intercept)
            best_svs_A = sv_A
            best_svs_B = sv_B

    # ---------------------------------------------------------------------
    # 2. DỰNG VIDEO CÓ CHÚ THÍCH (LEGEND)
    # ---------------------------------------------------------------------
    fig = plt.figure(figsize=(11, 8))
    gs = fig.add_gridspec(2, 1, height_ratios=[0.1, 0.9])

    # TẦNG TRÊN: BẢNG ĐIỆN TỬ
    ax_text = fig.add_subplot(gs[0])
    ax_text.axis('off')
    info_text = ax_text.text(0.5, 0.5, "", ha='center', va='center',
                             fontsize=13, fontweight='bold', color='#2c3e50',
                             bbox=dict(boxstyle="round,pad=0.4", facecolor="#f1f2f6", edgecolor="none"))

    # TẦNG DƯỚI: ĐỒ THỊ
    ax = fig.add_subplot(gs[1])

    # Vẽ dữ liệu thô với nhãn để hiện Legend
    ax.scatter(X_demo[y_demo == label_A][:, 0], X_demo[y_demo == label_A][:, 1], color='#e74c3c', s=80, alpha=0.4, label=label_A)
    ax.scatter(X_demo[y_demo == label_B][:, 0], X_demo[y_demo == label_B][:, 1], color='#3498db', s=80, alpha=0.4, label=label_B)

    # Khởi tạo các thành phần đồ họa sẽ thay đổi trong video
    line_main, = ax.plot([], [], color='#2ecc71', lw=3, zorder=3, label='Đường ranh giới')
    line_margin_up, = ax.plot([], [], color='gray', linestyle='--', lw=1.5, zorder=3, label='Đường lề (Margin)')
    line_margin_down, = ax.plot([], [], color='gray', linestyle='--', lw=1.5, zorder=3) # Không đặt label để tránh lặp trong legend

    # Support Vectors (Vòng tròn vàng)
    scatter_sv_A = ax.scatter([], [], facecolors='none', edgecolors='gold', s=300, linewidths=3, zorder=5, label='Support Vectors')
    scatter_sv_B = ax.scatter([], [], facecolors='none', edgecolors='gold', s=300, linewidths=3, zorder=5)

    ax.set_xlim(global_x_min - 0.05, global_x_max + 0.05)
    ax.set_ylim(global_y_min - 0.05, global_y_max + 0.05)
    ax.set_xlabel("Trục Contrast", fontweight='bold')
    ax.set_ylabel("Trục Entropy", fontweight='bold')
    ax.grid(True)

    # Hiển thị Legend cố định ở góc
    ax.legend(loc='upper right', framealpha=0.9)

    def init_step3():
        line_main.set_data([], [])
        line_margin_up.set_data([], [])
        line_margin_down.set_data([], [])
        scatter_sv_A.set_offsets(np.empty((0, 2)))
        scatter_sv_B.set_offsets(np.empty((0, 2)))
        info_text.set_text("")
        return line_main, line_margin_up, line_margin_down, scatter_sv_A, scatter_sv_B, info_text

    def animate_step3(i):
        m, intercept, margin, sv_A, sv_B = frames_step3[i]

        x_vals = np.array([global_x_min - 1, global_x_max + 1])
        y_vals = m * x_vals + intercept

        # Tính toán vạch kẻ Margin
        y_offset = margin * np.sqrt(m**2 + 1)
        y_vals_up = m * x_vals + (intercept + y_offset)
        y_vals_down = m * x_vals + (intercept - y_offset)

        line_main.set_data(x_vals, y_vals)
        line_margin_up.set_data(x_vals, y_vals_up)
        line_margin_down.set_data(x_vals, y_vals_down)

        scatter_sv_A.set_offsets(sv_A)
        scatter_sv_B.set_offsets(sv_B)

        if margin == best_margin:
            status = "-> [BEST] MAX MARGIN!"
            line_main.set_color('#8e44ad') # Màu tím cho đường tốt nhất
            line_main.set_label('Đường tối ưu (Best)') # Cập nhật nhãn nếu muốn (tùy chọn)
        else:
            status = "DANG TINH TOAN..."
            line_main.set_color('#2ecc71')

        info_str = f"Đường số {i+1}/{len(frames_step3)}  |  Độ rộng Lề: {margin:>6.4f}  |  {status}"
        info_text.set_text(info_str)

        return line_main, line_margin_up, line_margin_down, scatter_sv_A, scatter_sv_B, info_text

    anim_step3 = FuncAnimation(fig, animate_step3, init_func=init_step3, frames=len(frames_step3), interval=150, blit=True)
    plt.close(fig)
    display(HTML(anim_step3.to_jshtml()))

    # ---------------------------------------------------------------------
    # 3. CHỐT HẠ: VẼ BỨC ẢNH TĨNH TỐI ƯU
    # ---------------------------------------------------------------------
    best_m, best_intercept = best_candidate
    print("\n" + "="*70)
    print("🏆 GIAI ĐOẠN 5: ĐÃ TÌM RA ĐƯỜNG TỐI ƯU NHẤT")
    print("="*70)

    result_text = fr"""
### 🎉 Kết quả:
* **Phương trình:** $y = {best_m:.4f}x + {best_intercept:.4f}$
* **Max Margin:** **{best_margin:.4f}**
* **Support Vectors:** Các điểm khoanh tròn vàng là những điểm "quan trọng nhất" quyết định vị trí ranh giới này.
    """
    display(Markdown(result_text))

    fig3, ax3 = plt.subplots(figsize=(11, 7))
    # Vẽ lại dữ liệu thô
    ax3.scatter(X_demo[y_demo == label_A][:, 0], X_demo[y_demo == label_A][:, 1], color='#e74c3c', s=100, alpha=0.8, label=label_A)
    ax3.scatter(X_demo[y_demo == label_B][:, 0], X_demo[y_demo == label_B][:, 1], color='#3498db', s=100, alpha=0.8, label=label_B)

    # Vẽ đường Tối ưu (Tím)
    x_vals_static = np.array([global_x_min - 0.5, global_x_max + 0.5])
    y_vals_best = best_m * x_vals_static + best_intercept
    ax3.plot(x_vals_static, y_vals_best, color='#8e44ad', lw=4, label='Optimal Hyperplane (Max Margin)')

    # Vẽ lề
    y_offset_best = best_margin * np.sqrt(best_m**2 + 1)
    ax3.plot(x_vals_static, y_vals_best + y_offset_best, color='gray', linestyle='--', lw=2, label='Margin Boundary')
    ax3.plot(x_vals_static, y_vals_best - y_offset_best, color='gray', linestyle='--', lw=2)

    # Support Vectors
    ax3.scatter(best_svs_A[:, 0], best_svs_A[:, 1], facecolors='none', edgecolors='gold', s=400, linewidths=3, label='Support Vectors')
    ax3.scatter(best_svs_B[:, 0], best_svs_B[:, 1], facecolors='none', edgecolors='gold', s=400, linewidths=3)

    ax3.set_xlim(global_x_min - 0.05, global_x_max + 0.05)
    ax3.set_ylim(global_y_min - 0.05, global_y_max + 0.05)
    ax3.set_title("KẾT QUẢ SVM: RANH GIỚI TỐI ƯU VÀ SUPPORT VECTORS", fontsize=15, fontweight='bold')
    ax3.legend(loc='upper right')
    ax3.grid(True)
    plt.show()
```

## Bước 4: Tối ưu hóa Toán học - Chốt hạ Siêu phẳng biên tối đa (Maximum Margin)

Giai đoạn này, máy tính sẽ không "thử mù" các đường thẳng nữa, mà chính thức chuyển bài toán hình học thành một bài toán **Quy hoạch bậc hai (Quadratic Programming)** để tìm ra đáp án chính xác tuyệt đối.



Thuật toán sẽ tính toán để tìm ra bộ hai thông số cốt lõi: **$w$** (Vector trọng số - quyết định độ nghiêng/hướng của ranh giới) và **$b$** (Hệ số tự do - quyết định vị trí tịnh tiến của ranh giới).

### 1. Thuật toán tìm $w$ và $b$ để Lề lớn nhất như thế nào?

Trong không gian toán học, khoảng cách (bề rộng của Lề) giữa 2 mép vạch kẻ đi qua các Véc-tơ hỗ trợ được tính bằng công thức tỉ lệ nghịch với độ lớn của vector $w$:
$$Margin = \frac{2}{\|w\|}$$

Từ công thức này, SVM thiết lập một nguyên tắc tối ưu hóa vô cùng chặt chẽ bao gồm 2 yếu tố:
* **Mục tiêu chính (Minimize):** Để phân thức $\frac{2}{\|w\|}$ đạt giá trị **LỚN NHẤT** (Lề rộng nhất), thuật toán phải tìm cách điều chỉnh tham số sao cho $\|w\|$ đạt giá trị **NHỎ NHẤT**. (Trong thực tế tính toán, máy tính sẽ tối thiểu hóa biểu thức $\frac{1}{2}\|w\|^2$ để dễ đạo hàm).
* **Điều kiện ràng buộc (Constraints):** Trong lúc cố gắng làm cho lề rộng nhất có thể, $w$ và $b$ vẫn bắt buộc phải thỏa mãn điều kiện: **Không có bất kỳ điểm dữ liệu nào bị rơi vào giữa lòng đường**. Mọi điểm dữ liệu $(x_i, y_i)$ đều phải nằm đúng bên phần đường của mình và nằm ngoài (hoặc nằm ngay trên) mép lề. Biểu diễn bằng toán học:
$$y_i(w \cdot x_i + b) \ge 1$$

Giải quyết xong hệ phương trình này, SVM sẽ tự động "chốt" được giá trị $w$ và $b$ tối ưu cuối cùng, tạo ra **Siêu phẳng tối ưu (Optimal Hyperplane)**.

### 2. Ý nghĩa sâu xa của Lề tối đa (Maximum Margin)

Tại sao thuật toán phải vất vả tìm kiếm đường ranh giới có khoảng trống lớn nhất, thay vì chỉ cần một đường thẳng kẻ bừa miễn sao chia đúng 2 phe là được?

* **Khả năng Tổng quát hóa (Generalization):** Nếu ranh giới nằm quá sát các điểm dữ liệu (Lề hẹp), mô hình đang mắc hội chứng "học vẹt" (Overfitting) tập dữ liệu huấn luyện. Bằng cách ép Lề phải rộng nhất có thể, SVM đang tạo ra một vùng đệm an toàn trung lập. Mô hình sẽ học được "bản chất" của sự khác biệt giữa hai lớp dữ liệu thay vì ghi nhớ cục bộ từng điểm ảnh.
* **Sức chịu đựng nhiễu (Robustness):** Trong thực tế, dữ liệu kiểm thử (Testing Set) hoặc dữ liệu mới sau này đưa vào hệ thống sẽ hiếm khi hoàn hảo; chúng luôn chứa nhiễu (noise) hoặc sai số (ví dụ: chiếc lá hơi dị dạng, ánh sáng chụp ảnh thay đổi). Một ranh giới có Lề rộng đóng vai trò như một **"bộ giảm xóc"**. Nhờ khoảng trống lớn ở giữa, các điểm dữ liệu mới dù có bị lệch đôi chút so với quỹ đạo chuẩn vẫn sẽ rơi vào vùng an toàn và được phân loại chính xác, tránh hiện tượng đoán nhầm (misclassification).

```python
# =====================================================================
# BƯỚC 4: TỐI ƯU HÓA TOÁN HỌC & KIỂM TRA ĐIỀU KIỆN RÀNG BUỘC
# Cập nhật: Fix lỗi mất Support Vector (Kế thừa biến trực tiếp từ Bước 3)
# =====================================================================

if 'best_candidate' not in locals():
    print("⚠️ LỖI: Vui lòng chạy hoàn thiện Bước 3 trước khi chạy Bước 4!")
else:
    # --- CẤU HÌNH GIAO DIỆN MẶC ĐỊNH (ÉP NỀN TRẮNG) ---
    plt.style.use('default')
    plt.rcParams['figure.facecolor'] = 'white'
    plt.rcParams['axes.facecolor'] = 'white'
    plt.rcParams['savefig.facecolor'] = 'white'

    intro_step4 = r"""
#### 🧮 GIAI ĐOẠN 6: Chuyển đổi sang Ma trận Toán học & Kiểm chứng Ràng buộc
Thuật toán SVM không dùng $y = mx + c$, mà chuẩn hóa thành vector trọng số $w$ và hệ số $b$.
* Mọi điểm dữ liệu Phe B (Xanh) được gán nhãn $y_i = +1$.
* Mọi điểm dữ liệu Phe A (Đỏ) được gán nhãn $y_i = -1$.
* **Điều kiện ràng buộc thép (Hard Margin):** Không có điểm nào được nằm trong vùng lề. Tính toán học biểu diễn bằng: $y_i(w \cdot x_i + b) \ge 1$.
    """
    display(Markdown(intro_step4))

    # 1. DỊCH PHƯƠNG TRÌNH HÌNH HỌC SANG VECTOR TOÁN HỌC SVM
    best_m, best_intercept = best_candidate

    # Vector pháp tuyến ban đầu từ y = mx + c -> -mx + y - c = 0
    w_raw = np.array([-best_m, 1.0])
    b_raw = -best_intercept

    # Chuẩn hóa w thành vector đơn vị để tính khoảng cách tuyệt đối
    w_norm = np.linalg.norm(w_raw)
    w_unit = w_raw / w_norm
    b_unit = b_raw / w_norm

    # Đảm bảo phe B (Leaf 9) nằm ở chiều dương (+), phe A ở chiều âm (-)
    if np.dot(mean_B, w_unit) + b_unit < 0:
        w_unit = -w_unit
        b_unit = -b_unit

    # Tính khoảng cách thực tế từ mọi điểm đến Ranh giới trung tâm
    actual_distances = np.dot(X_demo, w_unit) + b_unit
    abs_distances = np.abs(actual_distances)
    optimal_margin = np.min(abs_distances) # Khớp 100% với best_margin ở Bước 3

    x_vals_static = np.array([global_x_min - 1, global_x_max + 1])
    y_vals_center_static = best_m * x_vals_static + best_intercept
    y_offset_static = optimal_margin * np.sqrt(best_m**2 + 1)

    # =====================================================================
    # [PHẦN 1] HOẠT ẢNH: QUÁ TRÌNH BƠM LỀ & KIỂM CHỨNG RÀNG BUỘC (DÙNG JSHTML)
    # =====================================================================
    print("\n⏳ Đang dựng Hoạt ảnh (Animation) quá trình Bơm phồng lề và quét lỗi ràng buộc...")

    inflation_ratios = np.concatenate([
        np.linspace(0.1, 1.0, 40),  # Phình dần đến mức tối ưu
        np.ones(15),                # Đứng yên tại mốc tối ưu để quan sát
        np.linspace(1.0, 1.6, 25)   # Cố tình phình to tiếp gây vi phạm
    ])

    fig4 = plt.figure(figsize=(12, 8))
    gs = fig4.add_gridspec(2, 1, height_ratios=[0.15, 0.85])

    ax_text4 = fig4.add_subplot(gs[0])
    ax_text4.axis('off')
    info_text4 = ax_text4.text(0.5, 0.5, "", ha='center', va='center',
                             fontsize=14, fontweight='bold', color='black',
                             bbox=dict(boxstyle="round,pad=0.5", facecolor="#f1f2f6", edgecolor="#34495e", lw=2))

    ax4 = fig4.add_subplot(gs[1])
    ax4.set_xlim(global_x_min - 0.05, global_x_max + 0.05)
    ax4.set_ylim(global_y_min - 0.05, global_y_max + 0.05)
    ax4.set_xlabel("Trục Contrast (Tọa độ X1)", fontweight='bold')
    ax4.set_ylabel("Trục Entropy (Tọa độ X2)", fontweight='bold')
    ax4.grid(True, linestyle=':', alpha=0.7)

    # Thêm Label cho các cụm điểm
    scatter_A = ax4.scatter(X_demo[y_demo == label_A][:, 0], X_demo[y_demo == label_A][:, 1], color='#e74c3c', s=80, alpha=0.6, label=label_A)
    scatter_B = ax4.scatter(X_demo[y_demo == label_B][:, 0], X_demo[y_demo == label_B][:, 1], color='#3498db', s=80, alpha=0.6, label=label_B)

    # Thêm Label cho các thành phần đường kẻ
    line_center, = ax4.plot([], [], color='#8e44ad', lw=3, zorder=3, label="Ranh giới trung tâm")
    line_m_up, = ax4.plot([], [], color='#2ecc71', linestyle='--', lw=2, zorder=3, label="Vùng lề (Margin)")
    line_m_down, = ax4.plot([], [], color='#2ecc71', linestyle='--', lw=2, zorder=3)

    # Thêm Label cho các điểm vi phạm
    scatter_violations = ax4.scatter([], [], facecolors='none', edgecolors='red', s=400, linewidths=4, zorder=6, label="Điểm vi phạm")

    # Kích hoạt Legend
    ax4.legend(loc='upper right', framealpha=0.9)

    def init_step4():
        line_center.set_data([], [])
        line_m_up.set_data([], [])
        line_m_down.set_data([], [])
        scatter_violations.set_offsets(np.empty((0, 2)))
        info_text4.set_text("")
        return line_center, line_m_up, line_m_down, scatter_violations, info_text4

    def animate_step4(i):
        ratio = inflation_ratios[i]
        current_margin = optimal_margin * ratio

        y_vals_center = best_m * x_vals_static + best_intercept
        y_offset = current_margin * np.sqrt(best_m**2 + 1)
        y_up = y_vals_center + y_offset
        y_down = y_vals_center - y_offset

        line_center.set_data(x_vals_static, y_vals_center)
        line_m_up.set_data(x_vals_static, y_up)
        line_m_down.set_data(x_vals_static, y_down)

        violators_mask = abs_distances < (current_margin - 1e-5)
        violating_points = X_demo[violators_mask]
        scatter_violations.set_offsets(violating_points)

        if ratio < 0.99:
            status = "DANG MO RONG LE (HOP LE)"
            color_text = "black"
            line_m_up.set_color('#34495e')
            line_m_down.set_color('#34495e')
        elif 0.99 <= ratio <= 1.01:
            status = "[TOI DA] DA CHAM SUPPORT VECTORS!"
            color_text = "green"
            line_m_up.set_color('#2ecc71')
            line_m_down.set_color('#2ecc71')
        else:
            status = f"VI PHAM RANG BUOC: {len(violating_points)} DIEM NAM TRONG LONG DUONG!"
            color_text = "red"
            line_m_up.set_color('#e74c3c')
            line_m_down.set_color('#e74c3c')

        info_str = f"Do rong Le hien tai: {current_margin:>6.4f} / Toi da: {optimal_margin:>6.4f}\n{status}"
        info_text4.set_text(info_str)
        info_text4.set_color(color_text)

        return line_center, line_m_up, line_m_down, scatter_violations, info_text4

    anim_step4 = FuncAnimation(fig4, animate_step4, init_func=init_step4, frames=len(inflation_ratios), interval=100, blit=True)
    plt.close(fig4)

    # Dùng JSHTML
    display(HTML(anim_step4.to_jshtml()))

    print("💡 BÀI HỌC CỐT LÕI TỪ HOẠT ẢNH TRÊN:")
    print("- Khi thuật toán tăng độ lớn của lề, không gian an toàn được mở rộng.")
    print("- Giới hạn cuối cùng (Max Margin) buộc phải dừng lại ngay khi lề chạm vào các Support Vectors.")
    print("- Nếu cố chấp mở rộng thêm, dữ liệu sẽ lọt vào 'vùng cấm' (khoanh đỏ), vi phạm toán học của Hard Margin SVM.")

    # =====================================================================
    # [PHẦN 2] ẢNH TĨNH: SO SÁNH TRỰC QUAN TRƯỚC VÀ SAU
    # =====================================================================
    print("\n" + "="*70)
    print("📸 SO SÁNH TRỰC QUAN: TRƯỚC VÀ SAU KHI ÁP DỤNG RÀNG BUỘC SVM")
    print("="*70)

    # Tạo 2 biểu đồ nằm cạnh nhau để dễ đối chiếu
    fig_compare, (ax_pre, ax_post) = plt.subplots(1, 2, figsize=(18, 7))

    # --- ĐỒ THỊ 1: TRƯỚC (Từ Bước 3) ---
    ax_pre.scatter(X_demo[y_demo == label_A][:, 0], X_demo[y_demo == label_A][:, 1], color='#e74c3c', s=80, alpha=0.6, label=label_A)
    ax_pre.scatter(X_demo[y_demo == label_B][:, 0], X_demo[y_demo == label_B][:, 1], color='#3498db', s=80, alpha=0.6, label=label_B)

    ax_pre.plot(x_vals_static, y_vals_center_static, color='#8e44ad', lw=3, label="Ranh giới (Bước 3)")
    ax_pre.plot(x_vals_static, y_vals_center_static + y_offset_static, color='gray', linestyle='--', lw=2, alpha=0.7, label="Lề biên thô")
    ax_pre.plot(x_vals_static, y_vals_center_static - y_offset_static, color='gray', linestyle='--', lw=2, alpha=0.7)

    ax_pre.set_xlim(global_x_min - 0.05, global_x_max + 0.05)
    ax_pre.set_ylim(global_y_min - 0.05, global_y_max + 0.05)
    ax_pre.set_title("TRƯỚC: Ranh giới thô (Chưa chốt Ràng buộc)", fontsize=13, fontweight='bold')
    ax_pre.set_xlabel("Trục Contrast", fontweight='bold')
    ax_pre.set_ylabel("Trục Entropy", fontweight='bold')
    ax_pre.legend(loc='upper right')
    ax_pre.grid(True, linestyle=':', alpha=0.7)

    # --- ĐỒ THỊ 2: SAU (Đã chốt hạ SVM) ---
    ax_post.scatter(X_demo[y_demo == label_A][:, 0], X_demo[y_demo == label_A][:, 1], color='#e74c3c', s=80, alpha=0.6, label=label_A)
    ax_post.scatter(X_demo[y_demo == label_B][:, 0], X_demo[y_demo == label_B][:, 1], color='#3498db', s=80, alpha=0.6, label=label_B)

    ax_post.plot(x_vals_static, y_vals_center_static, color='#8e44ad', lw=4, label="Siêu phẳng Biên Tối đa")
    ax_post.plot(x_vals_static, y_vals_center_static + y_offset_static, color='#2ecc71', linestyle='--', lw=2.5, label="Lề (+1)")
    ax_post.plot(x_vals_static, y_vals_center_static - y_offset_static, color='#2ecc71', linestyle='--', lw=2.5, label="Lề (-1)")

    # FIX TẠI ĐÂY: Sử dụng trực tiếp best_svs_A và best_svs_B được lưu từ Bước 3 thay vì tính lại bằng abs_distances
    ax_post.scatter(best_svs_A[:, 0], best_svs_A[:, 1], facecolors='none', edgecolors='gold', s=400, linewidths=3, zorder=6, label="Support Vectors")
    ax_post.scatter(best_svs_B[:, 0], best_svs_B[:, 1], facecolors='none', edgecolors='gold', s=400, linewidths=3, zorder=6)

    ax_post.set_xlim(global_x_min - 0.05, global_x_max + 0.05)
    ax_post.set_ylim(global_y_min - 0.05, global_y_max + 0.05)
    ax_post.set_title("SAU: Chốt hạ Siêu Phẳng (Thỏa mãn Hard Margin)", fontsize=13, fontweight='bold', color='#2c3e50')
    ax_post.set_xlabel("Trục Contrast", fontweight='bold')
    ax_post.set_ylabel("Trục Entropy", fontweight='bold')
    ax_post.legend(loc='upper right')
    ax_post.grid(True, linestyle=':', alpha=0.7)

    plt.tight_layout()
    plt.show()
```

## BƯỚC 5: Xử lý Dữ liệu Phi tuyến tính với "Kernel Trick"

Ở các bước trước, chúng ta đã chia tách dữ liệu bằng một đường thẳng. Nhưng trong thực tế, dữ liệu hiếm khi xếp gọn gàng thành 2 phe. Chuyện gì sẽ xảy ra nếu phe Đỏ nằm tụ tập ở giữa, và phe Xanh bao vây xung quanh tạo thành một vòng tròn? Một đường thẳng phẳng (Linear) lúc này hoàn toàn bất lực.

Để giải quyết bài toán này, SVM sử dụng một vũ khí tối thượng: **Kernel Trick**.

### 1. Ánh xạ chiều (Dimensionality Mapping) - "Bơm" dữ liệu lên không gian cao hơn
Nếu không thể cắt dữ liệu trên mặt phẳng 2D, tại sao không nhấc chúng lên không gian 3D?
* **Cách hoạt động:** Thuật toán tạo ra một trục tọa độ mới (ví dụ trục $Z$) dựa trên các đặc trưng hiện có. Các điểm dữ liệu đang nằm bẹp trên tờ giấy 2D sẽ được "bơm" bay lên không trung theo các độ cao khác nhau.
* **Kết quả:** Những điểm lúc nãy đan xen nhau giờ đây đã phân tầng rõ rệt về mặt độ cao. Lúc này, SVM chỉ việc nhẹ nhàng chèn một **mặt phẳng cứng (Hyperplane)** nằm ngang ở giữa để cắt đôi chúng. Khi chiếu mặt phẳng đó ngược về không gian 2D ban đầu, ta sẽ thấy một đường ranh giới uốn cong hoàn hảo (như đường tròn hoặc elip).

### 2. Tại sao gọi là "Trick" (Thủ thuật)?
Việc tính toán tọa độ mới cho hàng triệu điểm dữ liệu ở không gian 3D, 4D... hay thậm chí là vô hạn chiều sẽ vắt kiệt sức mạnh của bất kỳ siêu máy tính nào.
* **Bản chất của cú lừa:** Toán học của SVM phát hiện ra rằng, để tìm được Siêu phẳng tối ưu, nó **không cần biết tọa độ chính xác** của điểm dữ liệu ở không gian nhiều chiều. Nó chỉ cần biết **tích vô hướng (dot product)** – tức là mức độ tương đồng giữa 2 điểm đó.
* **Hàm Kernel ra đời:** Hàm Kernel tính toán trực tiếp khoảng cách của các điểm ở không gian vô hạn chiều ngay khi chúng vẫn đang nằm ở không gian 2D ban đầu. Nhờ "đi tắt" như vậy, máy tính giải quyết bài toán phức tạp mà không bị sập nguồn vì quá tải.

### 3. Hàm RBF (Radial Basis Function) - Vị vua của các loại Kernel
Có nhiều loại hàm Kernel (như Đa thức Polynomial, Sigmoid...), nhưng **RBF** là loại mạnh mẽ và được sử dụng mặc định nhiều nhất.

> **Cơ chế hoạt động của RBF:**
> RBF đánh giá sự tương đồng dựa trên khoảng cách. Điểm nào càng gần "tâm điểm" thì giá trị RBF càng cao, càng xa thì càng thấp (như ném một hòn đá xuống hồ và sóng gợn lan tỏa ra xa).

Tính toán học của RBF được biểu diễn bằng công thức:
$$K(x, x') = \exp(-\gamma ||x - x'||^2)$$

Trong đó:
* $||x - x'||^2$: Là khoảng cách hình học giữa 2 điểm dữ liệu.
* $\gamma$ (Gamma): Là một tham số điều chỉnh độ "rộng" của vùng ảnh hưởng.
    * Nếu $\gamma$ lớn: Ranh giới sẽ uốn lượn ngoằn ngoèo, ôm sát từng điểm dữ liệu (rất dễ bị Overfitting).
    * Nếu $\gamma$ nhỏ: Ranh giới sẽ mượt mà, bao quát thành các vùng rộng lớn.

**Tóm lại:** Hàm RBF biến các điểm dữ liệu thành các "đỉnh núi" và "thung lũng" trong không gian nhiều chiều. Nhờ đó, SVM có thể khoanh vùng các cụm dữ liệu đan xen một cách cực kỳ linh hoạt, tạo ra các ranh giới tự do hệt như những "hòn đảo" trên bản đồ.

```python
# =====================================================================
# BƯỚC 5: MÔ PHỎNG TRỰC QUAN KERNEL TRICK & RBF TRÊN DỮ LIỆU THỰC TẾ
# Kế thừa dữ liệu từ Bước 4 và tuân thủ đúng 4 quy trình lý thuyết
# =====================================================================

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML, display, Markdown
from sklearn.svm import SVC
import textwrap

if 'X_demo' not in locals():
    print("⚠️ LỖI: Vui lòng chạy dữ liệu từ các Bước trước đó để lấy X_demo và y_demo!")
else:
    # --- CẤU HÌNH GIAO DIỆN ---
    plt.style.use('default')
    plt.rcParams['figure.facecolor'] = 'white'

    # --- GIẢI THÍCH CƠ CHẾ TẠO TRỤC Z ---
    z_explanation = textwrap.dedent("""
    ### 💡 Chìa khóa của phép thuật: Trục Z (Độ cao) được tạo ra như thế nào?
    Để nhấc dữ liệu từ 2D lên 3D, chúng ta cần một trục tọa độ thứ 3 (trục Z). Trục này không phải do đo đạc thực tế, mà được **tự tạo ra bằng toán học** dựa trên 2 trục cũ.

    Trong ví dụ này, chúng ta dùng hàm ánh xạ đa thức (Polynomial Mapping) cơ bản:
    **Z = X1^2 + X2^2** (Tức là: `Contrast^2 + Entropy^2`)

    **Hiệu ứng xảy ra:**
    1. Những điểm Lá có giá trị Contrast và Entropy nhỏ (gần gốc tọa độ) sẽ có Z rất nhỏ $\\rightarrow$ **Nằm thấp sát mặt đất**.
    2. Những điểm Lá có giá trị lớn (xa gốc tọa độ) sẽ có Z bùng nổ lên rất lớn $\\rightarrow$ **Bị đẩy bay lên cao**.
    3. Sự chênh lệch này vô tình biến mặt phẳng 2D phẳng lì thành một hình "Cái Bát" 3D, khiến các điểm tự phân tầng độ cao!
    """)
    display(Markdown(z_explanation))

    # Chuẩn bị trục Z để tạo độ cao (Ánh xạ đa thức cơ bản: Z = X1^2 + X2^2)
    Z_demo = X_demo[:, 0]**2 + X_demo[:, 1]**2

    # =====================================================================
    # [PHẦN 1] HOẠT ẢNH: QUÁ TRÌNH "BƠM" DỮ LIỆU TỪ 2D LÊN 3D
    # =====================================================================
    print("⏳ Đang dựng Hoạt ảnh (Animation): Phép thuật bơm dữ liệu lên không gian 3D...")

    fig_anim = plt.figure(figsize=(10, 8))
    ax_anim = fig_anim.add_subplot(111, projection='3d')

    frames = 60

    def animate_3d(i):
        ax_anim.clear()

        # Tiến trình từ 0.0 đến 1.0
        progress = min(i / (frames * 0.8), 1.0)

        # Z tăng dần từ 0 lên giá trị thực
        current_Z = Z_demo * progress

        # Góc nhìn Camera (elev) chuyển mượt từ 90 độ (nhìn từ trên xuống như 2D) về 15 độ (nhìn ngang 3D)
        current_elev = 90 - (75 * progress)

        ax_anim.scatter(X_demo[y_demo == label_A][:, 0], X_demo[y_demo == label_A][:, 1], current_Z[y_demo == label_A], color='#e74c3c', s=70, edgecolors='w', label=label_A)
        ax_anim.scatter(X_demo[y_demo == label_B][:, 0], X_demo[y_demo == label_B][:, 1], current_Z[y_demo == label_B], color='#3498db', s=70, edgecolors='w', label=label_B)

        ax_anim.set_xlim(X_demo[:, 0].min()-0.05, X_demo[:, 0].max()+0.05)
        ax_anim.set_ylim(X_demo[:, 1].min()-0.05, X_demo[:, 1].max()+0.05)
        ax_anim.set_zlim(0, Z_demo.max() + 0.05)

        ax_anim.set_xlabel("Contrast")
        ax_anim.set_ylabel("Entropy")
        ax_anim.set_zlabel("Độ cao (Z)")
        ax_anim.set_title(f"Ánh xạ không gian: {int(progress*100)}% \n(Từ mặt phẳng bẹp 2D bay lên không gian 3D)", fontweight='bold', fontsize=12)

        # Khóa góc xoay ngang, chỉ lật dọc góc nhìn
        ax_anim.view_init(elev=current_elev, azim=-60)
        return []

    anim_3d = FuncAnimation(fig_anim, animate_3d, frames=frames, interval=80, blit=False)
    plt.close(fig_anim)
    display(HTML(anim_3d.to_jshtml()))

    # =====================================================================
    # [PHẦN 2] TRỰC QUAN HÓA 4 BƯỚC HOÀN CHỈNH
    # =====================================================================
    print("\n" + "="*70)
    print("📸 TỔNG HỢP 4 CÔNG ĐOẠN XỬ LÝ (THEO LÝ THUYẾT KERNEL TRICK)")
    print("="*70)

    fig_4steps = plt.figure(figsize=(20, 16))

    # --- ẢNH 1: TRẠNG THÁI 2D (Chưa ánh xạ) ---
    ax1 = fig_4steps.add_subplot(2, 2, 1)
    ax1.scatter(X_demo[y_demo == label_A][:, 0], X_demo[y_demo == label_A][:, 1], color='#e74c3c', s=60, alpha=0.8, label=label_A)
    ax1.scatter(X_demo[y_demo == label_B][:, 0], X_demo[y_demo == label_B][:, 1], color='#3498db', s=60, alpha=0.8, label=label_B)
    ax1.set_title("1. MẶT PHẲNG 2D: Dữ liệu đang nằm bẹp", fontsize=14, fontweight='bold')
    ax1.set_xlabel("Contrast")
    ax1.set_ylabel("Entropy")
    ax1.legend()
    ax1.grid(True, linestyle=':', alpha=0.6)

    # --- ẢNH 2: ÁNH XẠ 3D (Đã bơm lên cao) ---
    ax2 = fig_4steps.add_subplot(2, 2, 2, projection='3d')
    ax2.scatter(X_demo[y_demo == label_A][:, 0], X_demo[y_demo == label_A][:, 1], Z_demo[y_demo == label_A], color='#e74c3c', s=50, alpha=0.8)
    ax2.scatter(X_demo[y_demo == label_B][:, 0], X_demo[y_demo == label_B][:, 1], Z_demo[y_demo == label_B], color='#3498db', s=50, alpha=0.8)
    ax2.set_title("2. ÁNH XẠ 3D: Dữ liệu đã phân tầng độ cao", fontsize=14, fontweight='bold')
    ax2.set_xlabel("Contrast")
    ax2.set_ylabel("Entropy")
    ax2.set_zlabel("Trục Z mới")
    ax2.view_init(elev=15, azim=-60)

    # --- ẢNH 3: CẮT BẰNG SIÊU PHẲNG Ở 3D ---
    ax3 = fig_4steps.add_subplot(2, 2, 3, projection='3d')
    ax3.scatter(X_demo[y_demo == label_A][:, 0], X_demo[y_demo == label_A][:, 1], Z_demo[y_demo == label_A], color='#e74c3c', s=50, alpha=0.8)
    ax3.scatter(X_demo[y_demo == label_B][:, 0], X_demo[y_demo == label_B][:, 1], Z_demo[y_demo == label_B], color='#3498db', s=50, alpha=0.8)

    # Tạo mặt phẳng cắt ngang
    xx_3d, yy_3d = np.meshgrid(np.linspace(X_demo[:, 0].min(), X_demo[:, 0].max(), 10),
                               np.linspace(X_demo[:, 1].min(), X_demo[:, 1].max(), 10))
    # Tìm mức cắt hợp lý (nằm giữa trung bình của 2 cụm theo trục Z)
    z_cut = (np.mean(Z_demo[y_demo == label_A]) + np.mean(Z_demo[y_demo == label_B])) / 2
    zz_3d = np.full(xx_3d.shape, z_cut)
    ax3.plot_surface(xx_3d, yy_3d, zz_3d, alpha=0.4, color='#2ecc71')

    ax3.set_title("3. GIẢI QUYẾT Ở 3D: Chèn Siêu Phẳng cắt ngang", fontsize=14, fontweight='bold')
    ax3.set_xlabel("Contrast")
    ax3.set_ylabel("Entropy")
    ax3.set_zlabel("Trục Z mới")
    ax3.view_init(elev=15, azim=-60)

    # --- ẢNH 4: TRẢ VỀ 2D (Mô phỏng hàm RBF) ---
    ax4 = fig_4steps.add_subplot(2, 2, 4)
    # Huấn luyện mô hình SVM thực thụ với Kernel RBF (Tự động hóa hoàn toàn việc ánh xạ)
    svm_rbf = SVC(kernel='rbf', C=1.0, gamma='scale')
    svm_rbf.fit(X_demo, y_demo)

    xx_grid, yy_grid = np.meshgrid(np.linspace(X_demo[:, 0].min()-0.05, X_demo[:, 0].max()+0.05, 300),
                                   np.linspace(X_demo[:, 1].min()-0.05, X_demo[:, 1].max()+0.05, 300))
    Z_rbf = svm_rbf.decision_function(np.c_[xx_grid.ravel(), yy_grid.ravel()])
    Z_rbf = Z_rbf.reshape(xx_grid.shape)

    ax4.scatter(X_demo[y_demo == label_A][:, 0], X_demo[y_demo == label_A][:, 1], color='#e74c3c', s=60, edgecolors='k', alpha=0.8)
    ax4.scatter(X_demo[y_demo == label_B][:, 0], X_demo[y_demo == label_B][:, 1], color='#3498db', s=60, edgecolors='k', alpha=0.8)

    # Ranh giới chính RBF
    ax4.contour(xx_grid, yy_grid, Z_rbf, levels=[0], linewidths=4, colors='#8e44ad')
    # Background Contour (Vùng ảnh hưởng)
    ax4.contourf(xx_grid, yy_grid, Z_rbf, levels=np.linspace(Z_rbf.min(), 0, 7), cmap='Reds', alpha=0.1)
    ax4.contourf(xx_grid, yy_grid, Z_rbf, levels=np.linspace(0, Z_rbf.max(), 7), cmap='Blues', alpha=0.1)

    ax4.set_title("4. TRẢ VỀ 2D (RBF): Ranh giới uyển chuyển, khoanh vùng thông minh", fontsize=14, fontweight='bold')
    ax4.set_xlabel("Contrast")
    ax4.set_ylabel("Entropy")
    ax4.grid(True, linestyle=':', alpha=0.6)

    plt.tight_layout(pad=4.0)
    plt.show()
```

## Bước 5: Xử lý Dữ liệu Phi tuyến tính với "Kernel Trick"

Trong thực tế, dữ liệu thường đan xen lộn xộn và không thể chia tách bằng một đường thẳng phẳng.
* **Ánh xạ chiều:** SVM sử dụng các hàm toán học để đẩy dữ liệu lên một không gian có số chiều cao hơn (ví dụ từ 2D lên 3D). Tại đây, các điểm dữ liệu vốn dĩ đan xen sẽ tách rời nhau ra, cho phép một mặt phẳng cắt ngang qua dễ dàng.
* **Hàm RBF (Radial Basis Function):** Đây là hàm Kernel phổ biến nhất, giúp SVM có thể tạo ra các đường ranh giới uốn lượn phức tạp xung quanh các cụm dữ liệu mà không tốn quá nhiều tài nguyên tính toán.



## Bước 6: Đánh giá và Xây dựng Đường ống Suy luận (Inference Pipeline)

Đây là bước cuối cùng để đưa mô hình từ lý thuyết vào ứng dụng thực tiễn.
1.  **Đánh giá hiệu năng:** Mô hình học từ tập huấn luyện sẽ được đem đi "thi" trên **Tập kiểm thử**. Dựa vào kết quả này, ta tính toán các chỉ số như Accuracy, Precision, Recall và đặc biệt là **F1-Score** để đảm bảo mô hình không bị thiên vị.
2.  **Suy luận (Inference):** Sau khi đạt yêu cầu, mô hình được đóng gói thành một "đường ống" tự động. Khi có một ảnh mới:
    * *Ảnh $\rightarrow$ Trích GLCM $\rightarrow$ Chuẩn hóa Min-Max $\rightarrow$ Áp dụng phương trình SVM $\rightarrow$ Kết quả phân loại.*
    * Quá trình này diễn ra cực nhanh, cho phép ứng dụng trong các hệ thống nhận diện thời gian thực.
